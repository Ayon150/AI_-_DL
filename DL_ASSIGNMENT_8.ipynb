```python
# ============================================================
# TASK-1 : YOLOv8 CUSTOM OBJECT DETECTION
# COMPLETE GOOGLE COLAB CODE
# (FOR DATASET WITH TRAIN / VALID / TEST ALREADY PRESENT)
# ============================================================


# ============================================================
# STEP 1 : CHECK GPU
# ============================================================

!nvidia-smi


# ============================================================
# STEP 2 : INSTALL LIBRARIES
# ============================================================

!pip install -q ultralytics


# ============================================================
# STEP 3 : IMPORT LIBRARIES
# ============================================================

from ultralytics import YOLO
from google.colab import files

import zipfile
import os
import cv2
import matplotlib.pyplot as plt


# ============================================================
# STEP 4 : UPLOAD DATASET ZIP
# ============================================================

print("Upload your YOLOv8 dataset ZIP file")

uploaded = files.upload()


# ============================================================
# STEP 5 : EXTRACT DATASET
# ============================================================

zip_file = list(uploaded.keys())[0]

extract_path = "/content/dataset"

with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset Extracted Successfully")


# ============================================================
# STEP 6 : SHOW DATASET STRUCTURE
# ============================================================

for root, dirs, files_in_dir in os.walk(extract_path):

    print("------------------------------------------------")
    print(root)

    print(files_in_dir[:3])


# ============================================================
# STEP 7 : DEFINE PATHS
# ============================================================

yaml_path = "/content/dataset/data.yaml"

train_images_path = "/content/dataset/train/images"

valid_images_path = "/content/dataset/valid/images"

test_images_path = "/content/dataset/test/images"


# ============================================================
# STEP 8 : CHECK PATHS
# ============================================================

print("\n================================")

print("DATA.YAML PATH")

print("================================")

print(yaml_path)

print("\n================================")

print("TRAIN IMAGE PATH")

print("================================")

print(train_images_path)

print("\n================================")

print("VALID IMAGE PATH")

print("================================")

print(valid_images_path)

print("\n================================")

print("TEST IMAGE PATH")

print("================================")

print(test_images_path)


# ============================================================
# STEP 9 : DISPLAY data.yaml
# ============================================================

print("\n================================")

print("DATA.YAML CONTENT")

print("================================")

with open(yaml_path, "r") as f:

    print(f.read())


# ============================================================
# STEP 10 : SHOW SAMPLE TRAIN IMAGE
# ============================================================

sample_image = os.listdir(train_images_path)[0]

sample_image_path = os.path.join(train_images_path, sample_image)

img = cv2.imread(sample_image_path)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))

plt.imshow(img)

plt.title("Sample Training Image")

plt.axis("off")

plt.show()


# ============================================================
# STEP 11 : LOAD PRETRAINED YOLOv8 MODEL
# ============================================================

model = YOLO("yolov8n.pt")


# ============================================================
# STEP 12 : TRAIN MODEL
# ============================================================

results = model.train(

    data=yaml_path,

    epochs=50,

    imgsz=640,

    batch=16,

    device=0,

    workers=2,

    project="Task1_Object_Detection",

    name="YOLOv8_Custom",

    save=True,

    # ----------------------------------------
    # DATA AUGMENTATION
    # ----------------------------------------

    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,

    degrees=10,

    translate=0.1,

    scale=0.5,

    shear=2,

    fliplr=0.5,

    mosaic=1.0,

    mixup=0.2,

    patience=20
)


# ============================================================
# STEP 13 : VALIDATE MODEL
# ============================================================

metrics = model.val()

print("\n================================")

print("VALIDATION RESULTS")

print("================================")

print(f"Precision : {metrics.box.mp:.4f}")

print(f"Recall    : {metrics.box.mr:.4f}")

print(f"mAP50     : {metrics.box.map50:.4f}")

print(f"mAP50-95  : {metrics.box.map:.4f}")


# ============================================================
# STEP 14 : SHOW TRAINING RESULT GRAPH
# ============================================================

results_img_path = "/content/runs/detect/Task1_Object_Detection/YOLOv8_Custom/results.png"

img = cv2.imread(results_img_path)

if img is not None:

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(15,10))

    plt.imshow(img)

    plt.title("Training Results")

    plt.axis("off")

    plt.show()

else:

    print("results.png not found")


# ============================================================
# STEP 15 : SHOW CONFUSION MATRIX
# ============================================================

conf_matrix_path = "/content/runs/detect/Task1_Object_Detection/YOLOv8_Custom/confusion_matrix.png"

img = cv2.imread(conf_matrix_path)

if img is not None:

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(10,10))

    plt.imshow(img)

    plt.title("Confusion Matrix")

    plt.axis("off")

    plt.show()

else:

    print("confusion_matrix.png not found")


# ============================================================
# STEP 16 : UPLOAD CUSTOM IMAGE FOR TESTING
# ============================================================

print("\nUpload a custom image for testing")

uploaded_test = files.upload()

test_image = list(uploaded_test.keys())[0]

print("Uploaded Image:", test_image)


# ============================================================
# STEP 17 : LOAD BEST TRAINED MODEL
# ============================================================

best_model_path = "/content/runs/detect/Task1_Object_Detection/YOLOv8_Custom/weights/best.pt"

trained_model = YOLO(best_model_path)


# ============================================================
# STEP 18 : RUN OBJECT DETECTION
# ============================================================

results = trained_model.predict(

    source=test_image,

    conf=0.25,

    save=True
)


# ============================================================
# STEP 19 : DISPLAY DETECTION RESULT
# ============================================================

save_dir = results[0].save_dir

predicted_images = os.listdir(save_dir)

pred_img_path = os.path.join(save_dir, predicted_images[0])

pred_img = cv2.imread(pred_img_path)

pred_img = cv2.cvtColor(pred_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,12))

plt.imshow(pred_img)

plt.title("Detection Result")

plt.axis("off")

plt.show()


# ============================================================
# STEP 20 : VIDEO DETECTION
# ============================================================

print("\nUpload a video for detection")

uploaded_video = files.upload()

video_file = list(uploaded_video.keys())[0]

video_results = trained_model.predict(

    source=video_file,

    conf=0.25,

    save=True
)

print("Video Detection Completed")


# ============================================================
# STEP 21 : TEST SET EVALUATION
# ============================================================

test_metrics = trained_model.val(

    data=yaml_path,

    split="test"
)

print("\n================================")

print("TEST SET RESULTS")

print("================================")

print(f"Precision : {test_metrics.box.mp:.4f}")

print(f"Recall    : {test_metrics.box.mr:.4f}")

print(f"mAP50     : {test_metrics.box.map50:.4f}")

print(f"mAP50-95  : {test_metrics.box.map:.4f}")


# ============================================================
# STEP 22 : INFERENCE SPEED TEST
# ============================================================

print("\n================================")

print("INFERENCE SPEED TEST")

print("================================")

trained_model.predict(

    source=test_image,

    conf=0.25,

    verbose=True
)


# ============================================================
# STEP 23 : COMPARE WITH ORIGINAL YOLOv8
# ============================================================

print("\n================================")

print("ORIGINAL PRETRAINED YOLOv8")

print("================================")

original_model = YOLO("yolov8n.pt")

original_model.predict(

    source=test_image,

    conf=0.25,

    save=True
)

print("Original YOLOv8 Prediction Done")


# ============================================================
# STEP 24 : SAVE MODEL TO GOOGLE DRIVE
# ============================================================

from google.colab import drive

drive.mount('/content/drive')

!cp -r /content/runs/detect/Task1_Object_Detection /content/drive/MyDrive/


# ============================================================
# STEP 25 : DOWNLOAD BEST MODEL
# ============================================================

files.download(best_model_path)


# ============================================================
# STEP 26 : EXPORT MODEL
# ============================================================

trained_model.export(format="onnx")

print("ONNX Model Exported")


# ============================================================
# STEP 27 : FINAL SUMMARY
# ============================================================

print("\n================================================")

print("TASK-1 COMPLETED SUCCESSFULLY")

print("================================================")

print("YOLOv8 Fine-Tuning Completed")

print("Custom Dataset Training Completed")

print("Validation Completed")

print("Testing Completed")

print("Image Detection Completed")

print("Video Detection Completed")

print("Inference Speed Measured")

print("Model Saved Successfully")

print("Ready For Report Writing")
```
